```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.float_format', "{:.2f}".format)
pd.set_option('display.max_columns', None)

from classes import Paths
```

```python
paths = Paths()
rides_info = pd.read_csv(paths.rides_info)
rides_info
driver_info = pd.read_csv(paths.driver_info)
driver_info
car_train = pd.read_csv(paths.car_train)
car_train

df = rides_info.merge(driver_info, on='user_id', how='left').merge(car_train, on='car_id', how='left')
df
```

```python
params = {'data' : df,
          'kind' : 'scatter',
          'x' : 'ride_date',
          'y' : 'user_time_accident',
          'row' : 'fuel_type',
          'size' : 'car_type',
          'hue' : 'target_class',
          'aspect' : 4, 
          'alpha' : 0.5}

g = sns.relplot(**params)
plt.xticks(rotation=45);
```
