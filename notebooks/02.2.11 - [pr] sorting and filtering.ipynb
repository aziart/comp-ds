```python
from classes import Paths

import numpy as np
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.precision", 2)
```

```python
paths = Paths()
path = paths.car_train
df = pd.read_csv(path)
df.head()
```

```python
sort_by_list = ['car_rating', 'riders']
cols2drop = ['random_name_1', 'target_reg', 'target_class', 'random_name_2']
columns_order_list = ['model', 'car_type', 'car_id', 'car_rating', 'year_to_start', 'year_to_work', 'riders', 'fuel_type']
critical_car_rating = 2
```

- Sort the rows first in descending order of the 1st column from the `sort_by_list` list, then in ascending order of the 2nd column from the list.
- Remove columns from the `cols2drop` list, taking into account that some columns may not be in `df`.
- Make the order of the columns in `df` according to the variable `columns_order_list: List[str]`.
- Delete the lines where `car_rating` < `critical_car_rating`.
- Delete the lines where the number of `riders` is greater than `10_000`, or the `car_type` machine type refers to either `standart` or `economy`.

```python
# Sort the rows first in descending order of the 1st column from the `sort_by_list` list, then in ascending order of the 2nd column from the list.

df.sort_values(by=sort_by_list
               , ascending=[False, True]
               , inplace=True)
df.head()
```

```python
# Remove columns from the `cols2drop` list, taking into account that some columns may not be in `df`.

display(cols2drop)
df.columns
for i in cols2drop:
    if i in df.columns:
        df.drop(i
                , axis=1
                , inplace=True)
df.head()
```

```python
# Make the order of the columns in `df` according to the variable `columns_order_list: List[str]`.

display(columns_order_list)
df = df[columns_order_list]
df
```

```python
# Delete the lines where `car_rating` < `critical_car_rating`.

mask = df['car_rating'] < critical_car_rating
df.drop(df.index[mask]
        , axis=0
        , inplace=True)
df.head()
```

```python
# Delete the lines where the number of `riders` is greater than `10_000`, or the `car_type` machine type refers to either `standart` or `economy`.

riders_max = 10000
excl_car_type = ['standart', 'economy']
mask = (df['riders'] > riders_max) | (df['car_type'].isin(excl_car_type))

df.drop(df[mask].index
        , inplace=True)
df
```
