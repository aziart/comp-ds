- Add a new row (with the index `new_index:(int / str)` and the values `new_data:list`).
- Delete the rows with indexes from the `del_index_list: list`.
- Increase the value of `destroy_degree` by 1 in a row with an index equal to `index`. The `index` variable contains the index of a row OR a number (index).
- Delete the `fix_date` column.
- Create a new efficient column that is equal to `destroy_degree` divided by `work_duration`.
- Square the `work_duration` column.
- Rename the `work_duration` column to `work_duration_square`.
- Delete duplicates in the `car_id` column.

```python
new_index = 50
new_data = ['new_car_id', 'new_worker_id', '2019-11-11 4:44','refuel_check', 3.5, 10]
del_index_list = [41, 18, 33, 7, 46]
index = 22
```

```python
df.head()
```

```python
# Add a new row (with the index `new_index:(int / str)` and the values `new_data:list`).

display(df.loc[new_index,:])
df.iloc[index,:] = new_data
df.iloc[index,:]
```

```python
# Delete the rows with indexes from the `del_index_list: list`.

df.drop(del_index_list
        , axis=0
        , inplace=True)
df
```

```python
# Increase the value of `destroy_degree` by 1 in a row with an index equal to `index`.
# The `index` variable contains the index of a row OR a number (index).

display(df.loc[index, 'destroy_degree'])
df.loc[index, 'destroy_degree'] = df.loc[index, 'destroy_degree'] +1
df.loc[index, 'destroy_degree']
```

```python
# Delete the `fix_date` column.

df.drop(['fix_date']
        , axis=1
        , inplace=True
        )
df
```

```python
# Create a new efficient column that is equal to `destroy_degree` divided by `work_duration`.

df['efficient'] = df['destroy_degree'] / df['work_duration']
df.head()
```

```python
# Square the `work_duration` column.

df['work_duration'] = df['work_duration'].apply(lambda x: x**2).astype('int64')
df
```

```python
# Rename the `work_duration` column to `work_duration_square`.

ren_cols = {
    "work_duration": "work_duration_square"
}
df.rename(columns=ren_cols
          , inplace=True)
df
```

```python
# Delete duplicates in the `car_id` column.
df.drop_duplicates(subset=['car_id']
                   , inplace=True)
df
```
