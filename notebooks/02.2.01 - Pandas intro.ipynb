### <center> Pandas </center>

```python
import pandas as pd # pandas: lib for tabular data
import numpy as np  # numpy: lib for math-l calculations
```

<div class="alert alert-info">
<b>Useful Information</b>


- __Delete rows or columns:__ `df.drop(name, axis=0, inplace=True)`  - `name` row index or column name. `axis=0` means, that we delete row, in order to delete columns instead of index we provide columns' names list and `axis=1` as arguments.

    
- __Drop duplicates:__ `df.drop_duplicates()` keep in mind `inplace=True`. If we want to drop duplicates by specific column, we provide `subset` argument with its name or list of names of specific columns.

    
    
- __Insert into table (specifying place where to insert):__ `df.insert(loc, column, value)`.  Here `loc` (where), `column` (name), `value` (what value). By default, columns are inserted into the end.

    
    
- __Rename column:__ `df.rename(columns = {'old_name':'new_name'}, inplace=True)`</div>

#### `Sorting DataFrame's`
<div class="alert alert-info">
    
`DataFrame`'s can be sorted. Let's see an example:</p>

```python
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
         'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
         'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
         'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

#create DataFrame from dictionary
df = pd.DataFrame(data, index=labels)
df
```

**By values**

```python
df.sort_values(by='age') 
```

**By index**

```python
df.sort_index(axis=1, ascending=False)
```

<div class="alert alert-info">
    
By default `ascending` parameter is set to be `True`. 
If we want to sort data by two or more columns, we can provide `by` and `ascending` parameters with lists of `column names` and `booleans` for each column `ascending`-value.
`inplace` parameter is set to be `False`, by default, too.

#### `Filtering DataFrames`

In order to filter DataFrame we can use logical expressions, which create CTE (Common Table Expression) with given conditions. [index, true/false values]

```python
df.visits>2
```

```python
df[df.visits>2]
```

If we want to set more than one condition, then every condition is taken into brackets `()`
- and is defined with `&`
- or is defined with `|`

```python
df.loc[(df.visits > 2) & (df.priority == 'yes')]
```

```python
df.loc[(df.visits > 2) | (df.age > 5)]
```

In order for us to output a part of the DataFrame with cheap prices (trip prices below 500), let's just try to write down this condition and output the part of the dataframe that meets this condition

```python
rule_cheap = rides_info.ride_cost < 500
rides_info[rule_cheap].head().round(2)
```

#### `Boolean Masks`

Also, one of the easiest ways to filter out the desired strings is to add a Boolean mask. Let the mask be based on the condition that the cost of the trip is greater than a certain value.

- Creating a Boolean mask `[True, False, True, ...]`
- Output the first 5 values of the strings that match the condition

```python
rule_cheap = rides_info['ride_cost'] < 426  
rides_info[rule_cheap].head()   
```

If we want to add a new column `cheap_price` - a characteristic of the cheap price of a trip: if the price in the first 25% apartment (that is, less than 426), then it will be a cheap price, set the value to `Yes`, otherwise `Nan`.

```python
rides_info.head()
```

```python
rule_cheap = rides_info['ride_cost'] < 426
display(rule_cheap)
rides_info.loc[rule_cheap, 'cheap_price'] = 'Yes'
rides_info.head().round(2)
```

#### `.filter()` and `.query()` methods

Retrieving rows or columns

```python
display(df.filter(items=['a', 'j'], axis=0))
display(df.filter(items=['age', 'priority'], axis=1))
```

<div class="alert alert-info">

__Explanation__

This method also has a `like` parameter. It selects rows/columns whose name has a substring assigned to it. The `items` and `like` parameters are mutually exclusive.

```python
display(df.head())
display(df.filter(like='pri', axis=1))
display(df.filter(like='pri', axis=1).join(df.filter(like='ag', axis=1)))
```

Alternative way of __filtering__

```python
df.query('age > visits')
```

<div class="alert alert-info">

__Explanation__

If we need to use a variable in our query, for example, `mean_age`, then we have to put `@` before it

`df.query('age > @mean_age')`

We can also provide our query conditions with more complicated logic using __'AND'__ (denoted with `&`) and __'OR'__ (denoted with `|`) operators. 

`df.query("age > @mean_age | animal == 'cat'")`




#### `.concat()`-method allows us concatenate pandas-objects along a specific axis.

```python
s1 = pd.Series(['a', 'b'])
s2 = pd.Series(['c', 'd'])
pd.concat([s1, s2])
```

```python
pd.concat([s1, s2], keys=['s1', 's2'],
          names=['Series name', 'Row ID'])
```

```python
pd.concat([s1, s2], axis=1, keys=['col1', 'col2'])
cdf = pd.concat([s1, s2], axis=1, keys=['col1', 'col2'], names=['index'])
display(cdf)
display("Columns:", cdf.columns)
display("Index", cdf.index)
```

`DataFrame`'s concatenation example

```python
df1 = pd.DataFrame([['a', 1], ['b', 2]],
                   columns=['letter', 'number'])
df2 = pd.DataFrame([['c', 3], ['d', 4]],
                   columns=['letter', 'number'])
display(df1)
display(df2)
```

There's also an opportunity to define indices and labels

```python
display(pd.concat([df1, df2], axis=0))
display(pd.concat([df1, df2], axis=0, keys=['df1', 'df2']))
display(pd.concat([df1, df2], axis=0, keys=['df1', 'df2'], names=['ind1', 'ind2']))
```

```python
cdf = pd.concat([df1, df2], axis=0, keys=['df1', 'df2'], names=['ind1', 'ind2'])
cdf.loc[[('df1', 0), ('df2', 0)], :]
```

```python
display(pd.concat([df1, df2], axis=1))
display(pd.concat([df1, df2], axis=1, keys=['df1', 'df2'], names=['ind1', 'ind2']))
```

```python
cdf = pd.concat([df1, df2], axis=1, keys=['df1', 'df2'], names=['ind1', 'ind2'])
cdf.loc[:, [('df1', 'letter'), ('df2', 'letter')]]
```
