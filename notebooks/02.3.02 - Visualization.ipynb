## Import Libraries

```python
import numpy as np
import pandas as pd
pd.set_option("display.max_columns", None)
pd.set_option("display.precision", 2)
import seaborn as sns

import warnings
warnings.simplefilter("ignore")

from classes import Paths
```

```python
paths = Paths()
rides_info = pd.read_csv(paths.rides_info)
car_train = pd.read_csv(paths.car_train)
driver_info = pd.read_csv(paths.driver_info)

display("rides_info", rides_info)
display("car_train", car_train)
display("driver_info", driver_info)
```

```python
df = car_train.merge(rides_info
                     , on="car_id"
                     , how="left")
df = df.merge(driver_info
              , on="user_id"
              , how="left")
df
```

## Distribution using `df.hist()`

<div class = "alert alert-success">

`df.hist()` - builds a histogram of the distribution of data across numeric columns

```python
df.hist(); # the bad way of doing it
```

<div class="alert alert-info">

It can be seen that the graphs turned out to be small and the textual information overlaps with each other, also the gender column was defined as numeric, although it is categorical. Let's fix this by adding the <code>figsize</code> parameter and dropping the unnecessary column out. Also, to avoid displaying service lines, you can add `None` or `;` at the end of the cell in front of the graphs.

```python
df.drop("sex", axis=1).hist(figsize=(20, 15)
                            , layout=(-1, 5));
```

## Distribution using `df.boxplot()`

<div class = "alert alert-success">

`df.boxplot()` - a function that allows us to draw so-called "boxes with whiskers" showing the median, standard deviations and the spread of a feature on a single graph. Let's visualize type of breakdown, depending on the driver's rating, and also change the <code>font size</code> parameter responsible for the font size.

```python
df.boxplot(
    column=['user_rating']
    , by="target_class"
    , fontsize=8
    , figsize=(15, 5)
);

df.boxplot(
    column=['car_rating']
    , by="target_class"
    , fontsize=8
    , figsize=(15, 5)
);

df.boxplot(
    column=['user_rating', 'car_rating']
    , by="target_class"
    , fontsize=8
    , figsize=(15, 5)
);
```

## Heat maps of the dependencies of numerical and categorical variables

### _.1 - Numerical Variables

```python
corr = df.corr(numeric_only=True).round(2)
corr.style.background_gradient(cmap="RdYlGn")
```

### _.2 - Categorical Variables

```python
pd.crosstab(
    df["target_class"],
    df["model"],
    margins=True,
    normalize=True
).style.background_gradient(cmap="RdYlGn")
```
