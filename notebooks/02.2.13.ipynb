- Add a `flag` column that is equal to `'good_line'` if `rating > 5 and speed_max < 100`. `'bad_line'` if `rating > 5 and speed_max > 100`. `'undefined'` in other cases.
- Add the `speed_check` column, which takes the values `'avg_more'` if `distance/ride_duration > speed_avg`, otherwise `'avg_less'`.
- Delete the columns from the `cols2drop` list.
- Delete the lines in which `car_id` has empty values (delete if and only if NA is there).
- Write the result to the `result` variable.

```python
cols2drop = ['ride_cost', 'stop_times', 'user_ride_quality']
```

```python
# Add a `flag` column that is equal to `'good_line'` if `rating > 5 and speed_max < 100`. `'bad_line'` if `rating > 5 and speed_max > 100`. `'undefined'` in other cases.

lf = (
    lambda x: 'good_line' 
      if (x.rating > 5 and x.speed_max < 100) 
      else ('bad_line' if x.rating > 5 and x.speed_max > 100 else 'undefined')
      )
df['flag'] = df.apply(lf, axis=1)
df
```

```python
# Add the `speed_check` column, which takes the values `'avg_more'` if `distance/ride_duration > speed_avg`, otherwise `'avg_less'`.

df.head()

lf = lambda x: 'avg_more' if x.distance / x.ride_duration > x.speed_avg else 'avg_less'
df['speed_check'] = df.apply(lf, axis=1)
df
```

```python
# Delete the columns from the `cols2drop` list.

display(cols2drop)
display(df.head())
for i in cols2drop:
    if i in df.columns:
        df.drop(i, axis=1, inplace=True)
display(df.head())
```

```python
# Delete the lines in which `car_id` has empty values (delete if and only if NA is there).

df.head()
mask = df.isna().car_id
df.drop(df.index[mask]
        , axis=0
        , inplace=True)
df
```

```python
# Write the result to the `result` variable.

result = df.copy()
result
```
