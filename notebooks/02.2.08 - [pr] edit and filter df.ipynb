```python
from classes import Paths
import pandas as pd
pd.set_option("display.precision", 2)
pd.set_option("display.max_columns", None)
```

```python
path = Paths.driver_info
df = pd.read_csv(path)
df
```

```python
tripple_accident = (df['user_time_accident'] * 3)
# display(tripple_accident, type(tripple_accident))

df.insert(loc=len(df.columns), column="tripple_accident", value=tripple_accident, allow_duplicates=True)
df

# df['tripple_accident'] = df['user_time_accident'] * 3
# df
```

```python
rides_on_rating = (df['user_rides'] / df['user_rating'])
df.insert(loc = len(df.columns),
          column = 'rides_on_rating',
          value=rides_on_rating,
          allow_duplicates=True)
df
```

```python
df = df[df['user_time_accident'] > 25]
df
```

```python
df = df[df['age'] > df['age'].mean()]
df
```
