```python
from classes import Paths
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.float_format', "{:.2f}".format)
pd.set_option('display.max_columns', None)
```

```python
paths = Paths()
path = paths.car_train
df = pd.read_csv(path)
df.head()
```

**1. The "wheel_shake" breakdown mainly occurred in the segment of "standart" type machines.**

<font color='red'>Answer: </font> False

```python
df.groupby('car_type').target_class.value_counts(normalize=True).unstack().T
```

**2. The "Nissan Qashqai" car model has the maximum number of "engine_ignition" class breakdowns.**

<font color='red'>Answer: </font> True

```python
# checking if Nissan Qashqai exists in dataset
qashqai_df = df[df['model'].str.contains('qashqai', case=False)]
qashqai_df
```

```python
model_target = df.groupby('model').target_class.value_counts(normalize=True).unstack()
model_target.loc[:,['engine_ignition']].sort_values(by='engine_ignition', ascending=False).head(10)
```

or this can be done using pd.crosstab

```python
pd.crosstab(df['model'], df['target_class'], normalize='index').sort_values(by='engine_ignition', ascending=False).head(10)
```

**3. The type of breakdown "wheel_shake" is less represented among the other car models.**

<font color='red'>Answer: </font> True

```python
df.target_class.value_counts(normalize=True)
```

**4. The second place in the number of breakdowns falls on premium type machines.**

<font color='red'>Answer: </font> False

```python
pd.crosstab(df['target_class']
            , df['car_type']
            , normalize=True
            , margins=True).T.sort_values(by='All', ascending=False).loc[:,['All']]
```

**5. The MINI CooperSE electric car has no "electro_bug" type breakdowns.**

<font color='red'>Answer: </font> True

```python
pd.crosstab(df['model']
          , df['target_class']
          , normalize='index').sort_values(by='electro_bug'
                                           , ascending=True).loc[:,['electro_bug']]
```

**6. The breakdown of the "engine_check" class is the most popular among all.**

<font color='red'>Answer: </font> False

```python
df.target_class.value_counts(normalize=True)*100
```
