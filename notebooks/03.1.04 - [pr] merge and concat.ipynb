Task:

The rides dataset contains the columns: car_id, ride_id, rating, ride_duration, ride_cost

The cars dataset contains the columns:car_id, model, riders, target_class

- Enrich the rides dataset about trips with information about the future type of breakdown using the cars dataset. 
- Delete the extra columns. The rides table should be supplemented only with the target_class column.
- Delete trips for which there is no information about the breakdown.  
- Write it to the result variable.

<img src="../imgs/03.1.04_4.png" width=600>

```python
cars = pd.read_csv(paths.merge_practice_sample_cars)
display(cars.head())
rides = pd.read_csv(paths.merge_practice_sample_rides)
display(rides.head())
```

```python
result = pd.merge(left=rides
                  , right=cars[['car_id','target_class']]
                  , left_on='car_id'
                  , right_on='car_id'
                  , how='left')

drop_mask = result['target_class'].isna()
result.drop(result[drop_mask].index, inplace=True)
result.head()
```

```python
result = pd.merge(left=rides
                  , right=cars[['car_id','target_class']]
                  , left_on='car_id'
                  , right_on='car_id'
                  , how='left')

result = result.dropna(subset=['target_class'])
result.head()
```
