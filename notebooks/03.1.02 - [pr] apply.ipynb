## 1. `.apply()`-method with 1 column

<div class="alert alert-success">

Using the `.apply()` method in Pandas, we can easily apply any functions to existing columns.  
For example, add a flag: "Was the speed of the trip higher than a certain value or not?"
</div>

```python
def my_awesome_function(x):
    if x > 30:
        return "high"
    else:
        return "low"
```

```python
df['speed_high_low'] = df['speed_avg'].apply(my_awesome_function)
df
```

```python
df.insert(loc=len(df.columns)
        , column='speed_high_low_2'
        , value=df['speed_avg'].apply(my_awesome_function)
        , allow_duplicates=True)
df.head()
```

```python
tmp = df.copy()
display(tmp.head(10))

for i in range(1,5+1):
    tmp.insert(loc=len(df.columns)
            , column='speed_high_low_2'+str(i)
            , value=tmp['speed_max'].apply(my_awesome_function)+"_"+str(i)
            , allow_duplicates=True)
tmp
```

<div class="alert alert-success">

To make the code more elegant, we can define functions using a lambda function.
</div>

```python
# Define lambda function with ternary operator inside
my_awesome_function = lambda x: 'more_30' if x > 30 else 'no_more_30'

# Apply function to column
df['speed_avg_greater_30'] = df['speed_avg'].apply(my_awesome_function)

# See the result
df['speed_avg_greater_30'].value_counts()
```

```python
df.insert(loc=len(df.columns)
        , column='speed_avg_greater_30_2'
        , value=df['speed_avg'].apply(my_awesome_function)
        , allow_duplicates=True)
df['speed_avg_greater_30'].value_counts()
```

## 2. `.apply()`-method with 2 and more columns

<div class="alert alert-success">

We can write a function to apply to multiple columns at once. Then we will be able to generate features depending on several features at once.  
In this case, it is important not to forget to specify which axis the function will be applied to in the `axis` parameter.
</div>

```python
def a_div_b(x):
    return x['speed_avg'] / x['speed_max']
df['avg_over_max'] = df.apply(a_div_b, axis=1)
df.head()
```

```python
df.insert(loc=len(df.columns)
          , column = 'avg_over_max_2'
          , value=df.apply(a_div_b, axis=1)
          , allow_duplicates=True)
df.head()
```

Create a new `final_ride_cost` column in `df`, which will depend on the `rating` and maximum speed (`speed_max`) of the trip. The values in this column will vary depending on two conditions:

If the rating is higher than 6 and the maximum speed of the trip is less than 120 km/h, then reduce the cost of the trip (`ride_cost`) by 5%.

Otherwise: increase the cost of the trip (`ride_ost`) by 5%.

In the new `final_ride_cost` column, discard the fractional part of the values, leaving only the integer part. Since the cost is measured in rubles, the fractional part does not matter.
</div>

```python
def calc_new_ride_cost(x):
    if x["rating"] > 6 and x["speed_max"] < 120:
        final_cost = x["ride_cost"] * 0.95
    else:
        final_cost = x["ride_cost"] * 1.05
    return int(final_cost)

df['final_ride_cost'] = df.apply(calc_new_ride_cost, axis=1)
df.head()
```
