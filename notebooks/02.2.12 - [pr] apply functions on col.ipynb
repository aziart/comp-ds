```python
paths = Paths()
path = paths.car_train
df = pd.read_csv(path)
df.head()
```

- Create a variable `mean_rating` with the average of the `car_rating` column.
- Increase by one all the values in the `car_rating` and `year_to_work` columns.
- Filter the cars with an above-average rating from the `car_rating` column in any convenient way (use the `mean_rating` variable).
- Convert the `year_to_work` column as follows: if the value is less than `2017`, then add `2`, otherwise divide the value entirely by `5`.
- Rename the `year_to_start` column to `new_year_to_start` and delete the `fuel_type` column.
- Add the `super_` prefix to all values from the `car_type` column (standart -> super_standart).
- Write the solution to the `result` variable.

```python
# Create a variable `mean_rating` with the average of the `car_rating` column.

mean_rating = df['car_rating'].mean()
mean_rating
```

```python
# Increase by one all the values in the `car_rating` and `year_to_work` columns.

display(df.head())
df.loc[:,['car_rating', 'year_to_work']] = df.loc[:, ['car_rating', 'year_to_work']].apply(lambda x: x+1)
df.head()

# df['car_rating'] = df['car_rating']+1
# df['year_to_work'] = df['year_to_work']+1
```

```python
# Filter the cars with an above-average rating from the `car_rating` column in any convenient way (use the `mean_rating` variable).

mask = df['car_rating'] > mean_rating
df = df.drop(df.index[~mask])
df
```

```python
# Convert the `year_to_work` column as follows: if the value is less than `2017`, then add `2`, otherwise divide the value entirely by `5`.

lf = (lambda x: x+2 if x<2017 else x/5)
df.loc[:,'year_to_work'] = df.loc[:,'year_to_work'].apply(lf).astype('int64')
df.head()
```

```python
# Rename the `year_to_start` column to `new_year_to_start` and delete the `fuel_type` column.

df.head()
rn = {'year_to_start': 'new_year_to_start'}
df.rename(columns=rn
          , inplace=True)
df.drop('fuel_type'
        , axis=1
        , inplace=True)
df.head()
```

```python
# Add the `super_` prefix to all values from the `car_type` column (standart -> super_standart).

df.loc[:, 'car_type'] = df.loc[:, 'car_type'].apply(lambda x: 'super_' + x)
df.head()
```

```python
# Write the solution to the `result` variable.
result = df.copy()
result.head()
```

```python
# YOUR CODE HERE
mean_rating = df['car_rating'].mean()

df.loc[:,['car_rating', 'year_to_work']] = df.loc[:, ['car_rating', 'year_to_work']].apply(lambda x: x+1)

mask = df['car_rating'] > mean_rating
df = df.drop(df.index[~mask])

df['year_to_work'] = df['year_to_work'].astype('float64')

lf = (lambda x: x+2 if x<2017 else x/5)
df.loc[:,'year_to_work'] = df.loc[:,'year_to_work'].apply(lf).astype('int64')

rn = {'year_to_start': 'new_year_to_start'}
df.rename(columns=rn
          , inplace=True)
df.drop('fuel_type'
        , axis=1
        , inplace=True)

df.loc[:, 'car_type'] = df.loc[:, 'car_type'].apply(lambda x: 'super_' + x)

result = df.copy()
```
