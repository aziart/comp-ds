```python
from classes import Paths

import numpy as np
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.precision", 2)
```

```python
paths = Paths()
path = paths.driver_info
df = pd.read_csv(path)
df.head()
```

<div class="alert alert-info">

- Count the number of NA's in each row and write this in a new column `nans_num`.
- Delete lines where the number of missed records is greater than `2`.
- Delete the lines where `age` has missing values.
- Fill in the missing values in the `user_time_accident` column with the median value.
- Leave only the columns `age`, `user_rides`, and `user_time_accident`.
- Copy final dataframe into the result variable.

```python
# Count the number of NA's in each row and write this in a new column `nans_num`.

nans_num = df.isna().sum(axis=1)
df.insert(loc=len(df.columns)
          , column="nans_num"
          , value=nans_num
          , allow_duplicates=True)
df[df['nans_num'] > 0]
```

```python
# Delete lines where the number of missed records is greater than `2`.

display(df.shape)

mask = df['nans_num'] > 2
df.drop(df.index[mask]
        , axis=0
        , inplace=True)
df
```

```python
# Delete the lines where `age` has missing values.

mask = df['age'].isna()
df.drop(df.index[mask]
        , axis=0
        , inplace=True)
df
```

```python
# Fill in the missing values in the `user_time_accident` column with the median value.

display(df['user_time_accident'].isna().sum()) # before filling

uta_median = df['user_time_accident'].median()
df.loc[:,['user_time_accident']] = df.loc[:,['user_time_accident']].fillna(uta_median)
df

display(df['user_time_accident'].isna().sum()) # after filling
```

```python
# Leave only the columns `age`, `user_rides`, and `user_time_accident`.

df.head()
leave_cols = ['age', 'user_rides', 'user_time_accident']
df = df.loc[:,leave_cols]
df
```

```python
# Copy final dataframe into the result variable.

result = df.copy()
```
